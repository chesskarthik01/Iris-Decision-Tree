# Iris Decision Tree #
In this project, we'll use sci-kit learn to generate a decision tree algorithm for the iris dataset.

```python
import pandas as pd
import numpy as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset("iris")
sns.set_style("darkgrid")
```

```python
iris.head()
```

```python
iris.info()
```

```python
iris.describe()
```

```python
sns.pairplot(iris,hue="species")
```

### Building the decision tree model ###
Begin with train-test split

```python
from sklearn.model_selection import train_test_split
```

```python
X = iris.drop("species",axis=1)
y=iris["species"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

### Visualising the decision tree using graphviz ###

```python
from IPython.display import Image
from sklearn.externals.six import StringIO
from sklearn.tree import export_graphviz
import pydot

features = list(iris.columns[:-1])
features
```

```python
dot_data = StringIO()
export_graphviz(dtree,out_file=dot_data,feature_names=features,filled=True,rounded=True)
graph = pydot.graph_from_dot_data(dot_data.getvalue())
Image(graph[0].create_png())
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test,predictions))
print("\n")
print(classification_report(y_test,predictions))
```

On this dataset our decision tree model performed very well, let's see if a random forest model performs any better

### Creating a random forest model ###

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train,y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,rfc_pred))
print("\n")
print(classification_report(y_test,rfc_pred))
```

<div class="alert-info">
In this case the random forest model had the same performance as our decision tree, maybe on a larger dataset the performance would've been different
